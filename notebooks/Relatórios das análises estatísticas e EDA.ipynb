## Análise do DataFrame de Aluguéis de Imóveis

### Descrição Geral

O DataFrame contém 48.894 entradas e 16 colunas, com dados sobre imóveis listados para aluguel. 
As principais colunas incluem informações sobre o imóvel (nome, preço, tipo de quarto), o anfitrião (nome, id), localização 
(bairro, latitude, longitude), e avaliações (número de avaliações, última avaliação, média de avaliações por mês).

### Estrutura do DataFrame

A seguir estão as colunas do DataFrame:

1. **id**: Identificador único do imóvel.
2. **nome**: Nome do imóvel.
3. **host_id**: Identificador único do anfitrião.
4. **host_name**: Nome do anfitrião.
5. **bairro_group**: Região maior onde o imóvel está localizado (ex: Manhattan, Brooklyn).
6. **bairro**: Bairro específico dentro da região maior.
7. **latitude**: Latitude geográfica do imóvel.
8. **longitude**: Longitude geográfica do imóvel.
9. **room_type**: Tipo de quarto disponível (ex: Entire home/apt, Private room).
10. **price**: Preço de aluguel por noite (em dólares).
11. **minimo_noites**: Número mínimo de noites necessárias para a reserva.
12. **numero_de_reviews**: Número de avaliações feitas para o imóvel.
13. **ultima_review**: Data da última avaliação do imóvel.
14. **reviews_por_mes**: Média de avaliações por mês.
15. **calculado_host_listings_count**: Número de imóveis listados pelo anfitrião.
16. **disponibilidade_365**: Número de dias disponíveis para aluguel durante o ano.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\USER\Documents\teste_indicium_precificacao.csv"
df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
```

```python
sns.histplot(df[df['price'] < 500]['price'], bins=50, kde=True, color="blue")
plt.title("Distribuição do Preço (limitado a $500 para melhor visualização)")
plt.xlabel("Preço ($)")
plt.ylabel("Frequência")

plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)
```

### Observações iniciais:
- A maioria dos preços está abaixo de $200.
- Existe uma cauda longa para valores mais altos, diminuindo a presença de outliers.
- Podemos investigar se bairros específicos influenciam esses valores elevados.

```python
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlação entre Variáveis Numéricas")

plt.show()
```

## observações importantes:

-  O preço (price) tem baixa correlação com todas as outras variáveis, indicando que o valor pode depender mais de fatores categóricos, como **bairro* e *tipo de quarto*.
- *Número de reviews e reviews por mês* têm uma correlação moderada (0.55), o que faz sentido, pois anúncios mais populares tendem a ter mais avaliações.
- *A longitude tem uma correlação negativa (-0.15) com o preço*, o que pode indicar uma tendência espacial nos valores dos imóveis.

# Minhas Hipóteses  

## A localização impacta diretamente no valor do dia a dia  
**Hipótese**: Acomodações em Manhattan possuem um preço médio mais alto do que Brooklyn, Queens, Bronx e Staten Island.  
**Oportunidade**: Os proprietários podem usar essa informação para ajustar os preços e destacar vantagens competitivas dos bairros mais baratos.  

## A duração mínima da estadia influencia a precificação  
**Hipótese**: Anúncios com uma exigência maior de noites mínimas tendem a ter preços mais baixos por dia.  
**Oportunidade**: Estratégias dinâmicas de precificação podem ser aplicadas para estadias mais longas, incentivando reservas com descontos progressivos.  

## O tipo de acomodação determina grande parte do preço  
**Hipótese**: Acomodações inteiras (Casa/apto inteiro) são, em média, mais caras do que quartos privados ou compartilhados.  
**Oportunidade**: Os proprietários podem decidir entre rentabilizar com volume (quartos privados mais baratos) ou com exclusividade (aluguel de espaços inteiros a preços premium).  

## A popularidade do anúncio pode considerar preços mais altos  
**Hipótese**: Anúncios com mais avaliações e uma avaliação média alta tendem a ter preços mais elevados.  
**Oportunidade**: Investir em boas avaliações, responder rapidamente aos clientes e melhorar a experiência pode gerar preços premium.  

## A disponibilidade afeta o preço das tarifas  
**Hipótese**: Anúncios disponíveis durante todo o ano podem ter um preço médio menor do que aqueles com disponibilidade limitada.  
**Oportunidade**: Estratégias de precificação dinâmicas podem ser aplicadas, aumentando os preços em épocas de alta demanda (como feriados e eventos locais).
